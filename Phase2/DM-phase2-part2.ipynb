```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
from efficient_apriori import apriori
```

```python
#import datasets
orders_dataset_raw = pd.read_csv("./digikala_dataset/orders.csv")
orders_dataset_raw.head(10)
```

```python
orders_dataset_raw["ID_Item"].value_counts()[0:35]
```

```python
orders_dataset = orders_dataset_raw[["ID_Customer","ID_Item"]]
```

```python
grouped = orders_dataset.groupby("ID_Customer")
transactions = [group["ID_Item"].values.tolist() for name, group in grouped]
transactions
```

```python
itemsets, rules = apriori(transactions, min_support=0.0006, min_confidence=0.00005)
```

```python
frequent_itemsets = dict(sorted(itemsets[1].items(), key=lambda item: item[1], reverse=True))
frequent_itemsets
```

```python

for rule in sorted(rules, key=lambda rule: rule.lift, reverse=True):
  print(rule)
```
